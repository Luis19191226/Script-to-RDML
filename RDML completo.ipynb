```python
import xml.etree.cElementTree as ET
rdml = ET.Element("rdml")

dateMade = ET.SubElement(rdml, "dateMade")
dateMade.text = "Fecha de ejecución"
dateUpdated = ET.SubElement(rdml, "dateUpdated")
dateUpdated.text = "Fecha de actualización"

#Información de ID: Las funciones de este modulo acceden al almacenamoento de los ID (idnetificadores) asignados por los repsositorios donde se almacenan los ficheros RDML
 
id = ET.SubElement(rdml, "id")

ET.SubElement(id, "publisher").text = "Editor"
ET.SubElement(id, "serialNumber").text = "Numero de serie"
ET.SubElement(id, "MD5Hash").text = "Hachis MD5"

# Información del experimentador: Las funciones de este modulo permiten extraer la información referente al experimentador o experimentadores que participan en el ensayo, asi como la información de la institución donde se ejecuta el mismo.

experimenter = ET.SubElement(rdml, "experimenter", id="ID00001")

ET.SubElement(experimenter, "firstName").text = "Nombre del experimentador"
ET.SubElement(experimenter, "lastName").text = "Apellido del experimentador"
ET.SubElement(experimenter, "email").text = "Correo del experimentador"
ET.SubElement(experimenter, "labName").text = "Nombre del laboratorio"
ET.SubElement(experimenter, "labAddress").text = "Correo del laboratorio"

# Información de la documentación: Las funciones de este modulo permiten almacenar la mayor parte de la información basada en las guías MIQE

documentation = ET.SubElement(rdml, "documentation")

ET.SubElement(documentation, "text").text = "Informacion de texto"

#Información de la muestra: Las funciones de este modulo permiten la extracción de la información acerca de las muestras y sus respectivas caracteristicas

sample = ET.SubElement(rdml, "sample")

description = ET.SubElement(sample, "description")
description.text = "string"
ET.SubElement(description, "documentation").text = "idReferencesType"
xRef = ET.SubElement(description, "xRef")
xRef.text = "xRefType"
ET.SubElement(xRef, "name").text = "string"
ET.SubElement(xRef, "id").text = "string"

ET.SubElement(sample, "type").text = "Apellido del experimentador"
ET.SubElement(sample, "interRunCalibrator").text = "Correo del experimentador"

quantify = ET.SubElement(sample, "quantify")
quantify.text = "Nombre del laboratorio"
ET.SubElement(quantify, "value").text = "valores flotantes"
ET.SubElement(quantify, "unit").text = "quantifyUnitType"

ET.SubElement(sample, "calibratorSample").text = "Correo del laboratorio"

cdnaSynthesisMethod = ET.SubElement(sample, "cdnaSynthesisMethod")
cdnaSynthesisMethod.text = "Nombre del experimentador"
ET.SubElement(cdnaSynthesisMethod, "enzyme").text = "Nombre de la Enzima"
ET.SubElement(cdnaSynthesisMethod, "primingMethod").text = "primingMethodType"
ET.SubElement(cdnaSynthesisMethod, "dnaseTreatment").text = "Tratamiento con dnasa"
ET.SubElement(cdnaSynthesisMethod, "thermalCyclingConditions").text = "Condiciones de termociclado"

ET.SubElement(sample, "templateRNAQuantify").text = "Doble"

templateRNAQualify = ET.SubElement(sample, "templateRNAQualify")
templateRNAQualify.text = "templateQualifyType"
ET.SubElement(templateRNAQualify, "method").text = "Metodo de Templado RNA"
ET.SubElement(templateRNAQualify, "result").text = "Resultado de Templado RNA"


ET.SubElement(sample, "templateDNAQuantify").text = "Doble"

templateDNAQualify = ET.SubElement(sample, "templateDNAQualify")
templateDNAQualify.text = "templateQualifyType"
ET.SubElement(templateDNAQualify, "method").text = "Metodo de Templado DNA"
ET.SubElement(templateDNAQualify, "result").text = "Resultado de Templado DNA"

#Información del objetivo: Las funciones de este modulo son utilizadas para recopilar la información sobre los genes y secuencias objetivo. 

target = ET.SubElement(rdml, "target")

description = ET.SubElement(target, "description")
description.text = "string"
ET.SubElement(description, "documentation").text = "idReferencesType"
xRef = ET.SubElement(description, "xRef")
xRef.text = "xRefType"
ET.SubElement(xRef, "name").text = "string"
ET.SubElement(xRef, "id").text = "string"

ET.SubElement(target, "type").text = "targetTypeType"
ET.SubElement(target, "amplificationEfficiency").text = "Eficiencia de amplificacion"
ET.SubElement(target, "detectionLimit").text = "Limite de deteccion"
ET.SubElement(target, "dyeId").text = "ID del colorante"

sequences = ET.SubElement(target, "sequences")
sequences.text = "Tipo de secuencias"
forwardPrimer = ET.SubElement(sequences, "forwardPrimer")
forwardPrimer.text = "Primer forward"
ET.SubElement(forwardPrimer, "threePrimeTag").text = "threePrimeTag"
ET.SubElement(forwardPrimer, "fivePrimeTag").text = "fivePrimeTag"
ET.SubElement(forwardPrimer, "sequence").text = "sequence"

ET.SubElement(sequences, "reversePrimer").text = "Primer reverse"
ET.SubElement(sequences, "probe1").text = "Prueba 1"
ET.SubElement(sequences, "probe2").text = "Prueba 2"

amplicon = ET.SubElement(sequences, "amplicon")
amplicon.text = "Amplicon"
ET.SubElement(amplicon, "threePrimeTag").text = "threePrimeTag"
ET.SubElement(amplicon, "fivePrimeTag").text = "fivePrimeTag"
ET.SubElement(amplicon, "sequence").text = "sequence"

commercialAssay = ET.SubElement(target, "commercialAssay")
commercialAssay.text = "Ensayo comercial"
ET.SubElement(commercialAssay, "company").text = "Nombre de la compania"
ET.SubElement(commercialAssay, "orderNumber").text = "Numero de orden"

# Información de las condiciones de ciclado termico: Las funciones de este modulo permite la extracción de información acerca de las condiciones de ciclado termico que son establecidas en el termociclador  

thermalCyclingConditions = ET.SubElement(rdml, "thermalCyclingConditions")

description = ET.SubElement(thermalCyclingConditions, "description")
description.text = "Condiciones de termociclado"
ET.SubElement(description, "documentation").text = "Documentacion del termociclado"

lidTemperature = ET.SubElement(thermalCyclingConditions, "lidTemperature")
lidTemperature.text = "Temperatura"
ET.SubElement(lidTemperature, "experimenter").text = "Experimentador"

step = ET.SubElement(lidTemperature, "step")
step.text = "Paso"

ET.SubElement(step, "nr").text = "Numero de pasos"
ET.SubElement(step, "description").text = "Descripcion"

temperature = ET.SubElement(step, "temperature")
temperature.text = "Temperatura"
ET.SubElement(temperature, "temperature").text = "Temperatura en float"
ET.SubElement(temperature, "duration").text = "Duracion de termociclado"
ET.SubElement(temperature, "temperatureChange").text = "Cambio de Temperatura"
ET.SubElement(temperature, "durationChange").text = "Cambio de duracion"
ET.SubElement(temperature, "measure").text = "Medida"
ET.SubElement(temperature, "ramp").text = "Rampa"

gradient = ET.SubElement(step, "gradient")
gradient.text = "Gradientes"
ET.SubElement(gradient, "highTemperature").text = "Temperatura alta"
ET.SubElement(gradient, "lowTemperature").text = "Temperatura baja"
ET.SubElement(gradient, "duration").text = "Duracion"
ET.SubElement(gradient, "temperatureChange").text = "Cambio de temperatura"
ET.SubElement(gradient, "durationChange").text = "Cambio de duracion"
ET.SubElement(gradient, "measure").text = "Medida"
ET.SubElement(gradient, "ramp").text = "Rampa"

loop = ET.SubElement(step, "loop")
loop.text = "Bucle"
ET.SubElement(loop, "goto").text = "Ir al elemento del arbol"
ET.SubElement(loop, "repeat").text = "Repetir"

pause = ET.SubElement(step, "pause")
pause.text = "Pausa"
ET.SubElement(pause, "temperature").text = "Temperatura"

ET.SubElement(step, "lidOpen").text = "Apertura de la tapa del termociclador"

# Información del experimento: Las funciones que se encuentran en este modulo perimten la recopilar los detalles experimentales, es decir, la información del diseño experimental ejecucado durante el ensayo 

experiment = ET.SubElement(rdml, "experiment")

description = ET.SubElement(experiment, "description")
description.text = "Descripcion del experimento"
ET.SubElement(description, "documentation").text = "Documentación del experimento"

run = ET.SubElement(experiment, "run")
run.text = "Información de la corrida"
description = ET.SubElement(run, "description")
description.text = "Descripción de la corrida"
ET.SubElement(description, "documentation").text = "Documentación de la corrida"
ET.SubElement(description, "experimenter").text = "Experimentador"

ET.SubElement(run, "instrument").text = "Instrumento para ejecutar el ensayo"

dataCollectionSoftware = ET.SubElement(run, "dataCollectionSoftware")
dataCollectionSoftware.text = "Datos extraidos del Software"
ET.SubElement(dataCollectionSoftware, "name").text = "Nombre del Software"
ET.SubElement(dataCollectionSoftware, "version").text = "Version del Software"

ET.SubElement(run, "backgroundDeterminationMethod").text = "Resumen del método de determinación"
ET.SubElement(run, "cqDetectionMethod").text = "Método de detección de Cq"
ET.SubElement(run, "thermalCyclingConditions").text = "Condiciones de termociclado del ensayo"
ET.SubElement(run, "pcrFormat").text = "Formato de la PCR"
ET.SubElement(run, "runDate").text = "Hora del ensayo"

react = ET.SubElement(run, "react")
react.text = "Reactivos"
ET.SubElement(react, "sample").text = "Muestra"

data = ET.SubElement(react, "data")
data.text = "Datos de la muestra"
ET.SubElement(data, "tar").text = "tar"
ET.SubElement(data, "cq").text = "Cq"

quantify = ET.SubElement(data, "quantify")
quantify.text = "Cuantificar"
ET.SubElement(quantify, "value").text = "Valor"
ET.SubElement(quantify, "unit").text = "Unidad"

excl = ET.SubElement(data, "excl")
excl.text = "string"
adp = ET.SubElement(excl, "adp")
adp.text = "Amplificacion de la curva"
ET.SubElement(adp, "cyc").text = "Cyc en float"
ET.SubElement(adp, "tmp").text = "Tmp en float"
ET.SubElement(adp, "fluor").text = "Fluor en float"
mdp = ET.SubElement(excl, "mdp")
mdp.text = "Curva de fusión"
ET.SubElement(mdp, "tmp").text = "Tmp en float"
ET.SubElement(mdp, "fluor").text = "Fluor en float"

ET.SubElement(data, "endPt").text = "Punto final"
ET.SubElement(data, "bgFluor").text = "BgFluor en float"
ET.SubElement(data, "quantFluor").text = "Cuantificación de fluor"

thirdPartyExtensions = ET.SubElement(rdml, "thirdPartyExtensions")
thirdPartyExtensions.text = "Extenciones"

arbol = ET.ElementTree(rdml)
arbol.write("example_rdml.xml")
```
